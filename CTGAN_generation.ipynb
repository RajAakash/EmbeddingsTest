```python
import pandas as pd
import numpy as np
from ctgan import CTGAN

# Example: let's create a numerical dataset
np.random.seed(42)
data = pd.read_csv('XSBenchOutput.csv')

print(data.head())

# Initialize CTGAN
ctgan = CTGAN(epochs=300)  # Increase epochs if dataset is large

# Since all columns are numerical, no need for categorical_columns
ctgan.fit(data)

# Generate 500 synthetic samples
synthetic_data = ctgan.sample(500)

print(synthetic_data.head())
synthetic_data.to_csv('XSBench_CTGAN.csv',index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in data.columns:
    plt.figure(figsize=(6,4))
    sns.kdeplot(data[column], label='Original', fill=True)
    sns.kdeplot(synthetic_data[column], label='Synthetic', fill=True)
    plt.title(f'Distribution of {column}')
    plt.legend()
    plt.show()
```
